```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("USA_Housing.csv")
```

```python
df.info()
```

# **X AND Y**

```python
df.columns
```

```python
x = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

# **EDA**

# **Train test split**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

# **Regression**

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
pred_y = lr.predict(x_test)
```

```python
b = x_test.iloc[0]
lr.score(x_test,y_test)
```

```python
sns.scatterplot(x=y_test,y=pred_y)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, pred_y))
print('MSE:', metrics.mean_squared_error(y_test, pred_y))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred_y)))
```

```python
print(r2_score(y_test,pred_y))
```
